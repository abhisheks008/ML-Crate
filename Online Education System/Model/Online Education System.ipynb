# Visualisation of Data

```python
plt.figure(figsize=(10, 10))
df.groupby('Age(Years)')['Performance in online'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(10, 10))
df.groupby('Number of Subjects')['Study time (Hours)'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(10, 10))
df.groupby('Age(Years)')['Clearing doubts with faculties in online mode'].mean().sort_values(ascending=False).plot(kind='bar')
```

# Visualisation of Data Using Pie Chart

```python
plt.figure(figsize=(10, 10))
top = df['Age(Years)'].value_counts().sort_values(ascending=False).head(8)
top.plot(kind='pie')
```

```python
plt.figure(figsize=(10, 10))
top = df['Average marks scored before pandemic in traditional classroom'].value_counts().sort_values(ascending=False).head(8)
top.plot(kind='pie')
```

```python
plt.figure(figsize=(10, 10))
top = df['Are you involved in any sports?'].value_counts().sort_values(ascending=False).head(8)
top.plot(kind='pie')
```
