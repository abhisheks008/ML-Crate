```python
from sklearn.preprocessing import LabelEncoder #To convert categorical data to numeric
from sklearn.model_selection import train_test_split #For data split

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
df1 = df.copy()
df1.drop(columns = ['institution','location','location code','ar rank','er rank','fsr rank','cpf rank','ifr rank','isr rank','irn rank','ger rank','score scaled'], inplace = True)
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['er score'] = df1['er score'].replace(np.nan, 0)
df1['er score'] = df1['er score'].replace(0,df['er score'].mean())
df1['er score'] = round(df1['er score'],1)

df1['fsr score'] = df1['fsr score'].replace(np.nan, 0)
df1['fsr score'] = df1['fsr score'].replace(0,df1['fsr score'].mean())
df1['fsr score'] = round(df1['fsr score'],1)

df1['cpf score'] = df1['cpf score'].replace(np.nan, 0)
df1['cpf score'] = df1['cpf score'].replace(0,df1['cpf score'].mean())
df1['cpf score'] = round(df1['cpf score'],1)

df1['ifr score'] = df1['ifr score'].replace(np.nan, 0)
df1['ifr score'] = df1['ifr score'].replace(0,df1['ifr score'].mean())
df1['ifr score'] = round(df1['ifr score'],1)

df1['isr score'] = df1['isr score'].replace(np.nan, 0)
df1['isr score'] = df1['isr score'].replace(0,df1['isr score'].mean())
df1['isr score'] = round(df1['isr score'],1)

df1['irn score'] = df1['irn score'].replace(np.nan, 0)
df1['irn score'] = df1['irn score'].replace(0,df1['irn score'].mean())
df1['irn score'] = round(df1['irn score'],1)

df1['ger score'] = df1['ger score'].replace(np.nan, 0)
df1['ger score'] = df1['ger score'].replace(0,df1['ger score'].mean())
df1['ger score'] = round(df1['ger score'],1)
```

# Model Creation
- Linear Regression
- Decision Tree
- Decision Tree

```python
train, test = train_test_split(df1, test_size=0.2, random_state=17)

print(f"No. of training examples: {train.shape[0]}")
print(f"No. of testing examples: {test.shape[0]}")
```

```python
x_train=train.drop('Rank',axis=1)
y_train=train[['Rank']]
```

```python
x_test=test.drop('Rank',axis=1)
y_test=test[['Rank']]
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
x_train.head(2)
```

```python
y_train.head(2)
```

# Linear Regression

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr_score = lr.score(x_test, y_test)
lr_score
```

```python
y_pred = lr.predict(x_test)
print(y_pred.flatten()[:5])
print(y_test.values.flatten()[:5])
```

```python
lr_mse = mean_squared_error(y_test.values, y_pred)
lr_rmse = mean_squared_error(y_test.values, y_pred,squared=False) #RMSE
lr_mae = mean_absolute_error(y_test.values, y_pred)
lr_r = r2_score(y_test.values, y_pred)

print('Score Mean Squared Error      = {}'.format(lr_mse))
print('Score Root Mean Squared Error = {}'.format(lr_rmse))
print('Score Mean Absolute Error     = {}'.format(lr_mae))
print('Score R2                      = {}'.format(lr_r))
```

```python
plt.figure(figsize=(15,10))
plt.plot(range(len(y_test)), y_test, marker='.', label="Actual")
plt.plot(range(len(y_test)), y_pred.flatten(), label="Prediction")
plt.ylabel('Rank')
plt.title('University ranking prediction with Linear Regression')
plt.legend()
plt.show()
```

```python
prediction = pd.DataFrame()
prediction['Actual'] = y_test.values.flatten()
prediction['Linear Regression'] = y_pred.flatten()
prediction['Linear Regression'] = prediction['Linear Regression'].round(0)
prediction.head()
```

# Decision Tree

```python
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
dt_score = dt.score(x_test, y_test)
dt_score
```

```python
y_pred = dt.predict(x_test)
print(y_pred.flatten()[:5])
print(y_test.values.flatten()[:5])
```

```python
dt_mse = mean_squared_error(y_test.values, y_pred)
dt_rmse = mean_squared_error(y_test.values, y_pred,squared=False) #RMSE
dt_mae = mean_absolute_error(y_test.values, y_pred)
dt_r = r2_score(y_test.values, y_pred)

print('Score Mean Squared Error      = {}'.format(dt_mse))
print('Score Root Mean Squared Error = {}'.format(dt_rmse))
print('Score Mean Absolute Error     = {}'.format(dt_mae))
print('Score R2                      = {}'.format(dt_r))
```

```python
plt.figure(figsize=(15,10))
plt.plot(range(len(y_test)), y_test, marker='.', label="Actual")
plt.plot(range(len(y_test)), y_pred.flatten(), label="Prediction")
plt.ylabel('Rank')
plt.title('University ranking prediction with DecisionTree')
plt.legend()
plt.show()
```

```python
prediction['DecisionTree'] = y_pred.flatten()
prediction['DecisionTree'] = prediction['DecisionTree'].round(0)
prediction.head(4)
```

# Random Forest

```python
plt.figure(figsize=(15,10))
plt.plot(range(len(y_test)), y_test, marker='.', label="Actual")
plt.plot(range(len(y_test)), y_pred.flatten(), label="Prediction")
plt.ylabel('Rank')
plt.title('University ranking prediction with RandomForest')
plt.legend()
plt.show()
```
