### Feature Scaling

```python
movies.iloc[0].genres  #the data is in the form of list of dictionaries
```

```python
#1.'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#2.we have to convert the format above in the format below
#3.['Action', 'Adventure', 'Fantasy', 'SciFi']
```

```python
#We can see that in the above code block line 1 is enclosed in string format, so we have to remove the string using a python function called ast
#this library ast will convert the string into list of dictionary
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#now you can see that the genres column is sorted out
```

```python
movies.iloc[0].keywords
```

```python
#we will do the same thing on the keywords column
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].cast
```

```python
#we need to take out the first 3 actor names
```

```python
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!=3:
        L.append(i['name'])
        counter+=1
    else:
       break
  return L
#first the value of counter will be 1, then the value of name will be appended, then the counter value will be 2 and so on till 3.
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
        L.append(i['name'])
        break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
#we have to convert the overview column from a list to a string
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
"""we have to remove the spaces between words like 'Sam Worthington' to 'SamWorthington' because if we do not
remove the space the recommender system will take sam as a different entity and worthington as a different entity so, 
whenever we type sam, we can get recommendations of other sams as well.
"""
```

```python
#removing spaces
movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
#as you can see in the first line of the output the space between Science and Fiction is removed
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
#we will make a new column called tags which will be the concatentation of overview, genres, keywords, cast and crew
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
#now we have to convert the list of the tags column in string format
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#we are doing stemming of words
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
#converting all the string in the tags column to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

## Vectorizing the text in Tags Column

```python
#we have to vectorize the text, and take the top 5000 words from the text
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english') #max_features=5000 means it takes the top most 5000 repeating words
#stop_words hyperparameter removes the stop words from the text
cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
#vector format of the first movie
vectors[0]
```

## Applying Cosine Similarity of Sklearn

```python
#we will calculate the cosine similarity(ranges from 0 to 1). Closer to 1 means more similar, closer to 0 means less similar
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity[0]
```

```python
list(enumerate(similarity[0])) #we hold the indexes of the movies
```

```python
#we need to take out the first five most common movies
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6] #gives the value of the similarity score in the descending order of the similarity score
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]  #fetches the index of the movie which we want to search
    distances = similarity[movie_index]  #gives the similarity values of the movie in that particular index
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6] #we get the first five common movies

    for i in movies_list:
        print(new_df.iloc[i[0]].title) #gives the title of the top 5 recommended movies according to the movie searched
```

```python
recommend('Avatar') #gives the top 5 recommended movies when we search for Avatar
```
